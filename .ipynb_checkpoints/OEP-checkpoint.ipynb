## Feature Analysis

#### Import modules

```python
import numpy as np
import pandas as pd
import missingno as msno
import seaborn as sn
import matplotlib.pyplot as plt
```

#### Get dataset

```python
data = pd.read_csv('winequality-white.csv',sep=';')
data.head()
```

#### Missing data detection

```python
msno.matrix(data,figsize=(10,3)).plot()
```

```python
msno.bar(data,figsize=(10,3)).plot()
```

#### There is no missing data in the  dataset

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
fig.set_size_inches(10,20)
sn.boxplot(data=data,orient="h",ax=axes[0])
sn.boxplot(data=data,y='quality',orient='pH',ax=axes[1])
```

```python
corrMatt = data.corr()
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
sn.heatmap(corrMatt, mask=mask,vmax=.8, square=True,annot=True)
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Adding extra Columnt
X = np.append(arr = np.ones((X.shape[0], 1)), values = X, axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Scaling the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Linear regressiong
regressor = LinearRegression()
regressor.fit(X_train, y_train)
predictions = regressor.predict(X_test)
```

```python
r2_score(y_test, predictions)
```

```python
# Backwards elimination
import statsmodels.formula.api as sm
X_opt = X[:, [0, 1, 2, 4, 6, 8, 9, 10, 11]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# Displaying the results
import matplotlib.pylab as plt
plt.scatter(y_test, predictions, c='g')
plt.xlabel('True Quality')
plt.ylabel('Predicted Quality')
plt.title('Predicted Quality Against True Quality ')
plt.show()
```
